```python
import numpy as np
import cv2
import time
import random
import math
import humanSkin
import convexHull
import threading


cap = cv2.VideoCapture("http://127.0.0.1:4747/mjpegfeed")
cap.set(cv2.CAP_PROP_FRAME_HEIGHT,480)
cap.set(cv2.CAP_PROP_FRAME_WIDTH,640)
hands=cv2.imread("hands.png")
hands=cv2.resize(hands,(640,300))
rules=cv2.imread("rules.png")
rules=cv2.resize(rules,(500,430))


scissors=cv2.imread("scissor.png")
scissors=cv2.resize(scissors,(150,150))


rock=cv2.imread("rock.png")
rock=cv2.resize(rock,(150,150))


paper=cv2.imread("paper.png")
paper=cv2.resize(paper,(150,150))


spock=cv2.imread("spock.png")
spock=cv2.resize(spock,(150,150))


lizard=cv2.imread("lizard.png")
lizard=cv2.resize(lizard,(150,150))


gestureImgs = {}

gestureImgs["scissors"] = scissors
gestureImgs["rock"] = rock
gestureImgs["paper"] = paper
gestureImgs["spock"] = spock
gestureImgs["lizard"] = lizard
gestureImgs[""] = newFrame=np.zeros((150,150,3), np.uint8)









start =time.time()
fCount=0
frameRate=0
stopThread=False
threadFrame=np.zeros((480, 640 , 3), np.uint8)

# font 
font = cv2.FONT_HERSHEY_SIMPLEX 

# org 
org = (50, 50) 

# org 
orgState = (50, 450) 

# org 
orgTitle = (700, 480) 
# org 
orgGesture = (700, 620) 

# org 
orgScore = (700, 720) 

# fontScale 
fontScale = 1

# Blue color in BGR 
color = (255, 255, 255) 

# Blue color in BGR 
colorPause = (0, 0, 255) 

# Blue color in BGR 
red = (0, 255, 255) 

# Line thickness of 2 px 
thickness = 2

# Using cv2.putText() method 

xOffset=500
yOffset=300


state="Wait Start Or End"

gestures=["rock", "paper", "scissors", "lizard", "spock"]
gesturesUser=""
gesturesCpu=""
whoWinString=""
userScore=0
cpuScore=0


    
def drawFrameRate():
    global fCount,frameRate,start,frame
    fCount=fCount+1
    end =time.time()
    if(end-start>=1): 
        frameRate=fCount
        fCount=0
        start =time.time()
    frame = cv2.putText(frame, str(frameRate)+" FPS", org, font, fontScale, color, thickness, cv2.LINE_AA)
    
    
def appendUI():
    global newFrame,frame,hands,rules,state
    
    if(state=="Pause"):
        frame = cv2.putText(frame, "State :"+ state+"       "+str(whoWinString), orgState , font, fontScale, colorPause, thickness, cv2.LINE_AA)
    else:
        frame = cv2.putText(frame, "State :"+ state+"       "+str(whoWinString), orgState , font, fontScale, color, thickness, cv2.LINE_AA)
    newFrame=np.zeros((frame.shape[0]+yOffset,(frame.shape[1]+xOffset),frame.shape[2]), np.uint8)
    newFrame[0:frame.shape[0],0:frame.shape[1],:]=frame
    newFrame[frame.shape[0]:frame.shape[0]+hands.shape[0]    ,    0:hands.shape[1]   ,:]=hands
    newFrame[   0:rules.shape[0]  ,    frame.shape[1]:frame.shape[1]+rules.shape[1]   ,:]=rules
    newFrame[   rules.shape[0]+100:rules.shape[0]+250,  frame.shape[1]+50:frame.shape[1]+200   ,:]=gestureImgs[gesturesUser]
    newFrame[   rules.shape[0]+100:rules.shape[0]+250,  frame.shape[1]+300:frame.shape[1]+450   ,:]=gestureImgs[gesturesCpu]


    newFrame = cv2.putText(newFrame, "USER     VS     CPU", orgTitle, font, fontScale, red, thickness, cv2.LINE_AA)
    #newFrame = cv2.putText(newFrame, str(gesturesUser)+"           "+str(gesturesCpu), orgGuesture, font, fontScale, red, thickness, cv2.LINE_AA)
    newFrame = cv2.putText(newFrame, str(userScore)+"               "+str(cpuScore), orgScore, font, fontScale, red, thickness, cv2.LINE_AA)


    
def whoWin():
    global gesturesUser,gesturesCpu,userScore,cpuScore,whoWinString
    if(gesturesUser==gesturesCpu):
        whoWinString="DRAW"
        
    if(gesturesUser=="rock"):
        if(gesturesCpu=="paper") or (gesturesCpu=="spock"):
            whoWinString="CPU WIN"
        elif(gesturesCpu=="scissors") or (gesturesCpu=="lizard"):
            whoWinString="USER WIN"

    elif(gesturesUser=="paper"):
        if(gesturesCpu=="rock") or (gesturesCpu=="spock"):
            whoWinString="USER WIN"
        elif(gesturesCpu=="scissors") or (gesturesCpu=="lizard"):
            whoWinString="CPU WIN"
         
    elif(gesturesUser=="scissors"):
        if(gesturesCpu=="rock") or (gesturesCpu=="spock"):
            whoWinString="CPU WIN"
        elif(gesturesCpu=="paper") or (gesturesCpu=="lizard"):
            whoWinString="USER WIN"
            
    elif(gesturesUser=="lizard"):
        if(gesturesCpu=="rock") or (gesturesCpu=="scissors"):
            whoWinString="CPU WIN"
        elif(gesturesCpu=="paper") or (gesturesCpu=="spock"):
            whoWinString="USER WIN"
            
    elif(gesturesUser=="spock"):
        if(gesturesCpu=="rock") or (gesturesCpu=="scissors"):
            whoWinString="USER WIN"
        elif(gesturesCpu=="lizard") or (gesturesCpu=="paper"):
            whoWinString="CPU WIN"      

    if(whoWinString=="USER WIN"):
        userScore=userScore+1
    elif(whoWinString=="CPU WIN"):
        cpuScore=cpuScore+1

def keepReadingFrames(cap):
    while True:
        global threadFrame,stopThread        
        retNow, frameNow=cap.read() 
        
        if retNow==True:
            threadFrame=np.copy(frameNow)

        if stopThread:
            break  
            
            



def main():

    global cap,frame,state,gesturesUser,gesturesCpu,userScore,cpuScore,gestures,whoWinString,stopThread,threadFrame
    count_paper = 0
    count_rock = 0
    count_scissors = 0
    count_lizard = 0
    count_spock = 0
    count_pause =0
    count_resume =0
    count_start =0
    count_end =0
    max_count=50
    t = threading.Thread(target=keepReadingFrames, name=keepReadingFrames, args=[cap])
    t.start() 
    while(cap.isOpened()):
        frame=np.copy(threadFrame)
        
        key="None"  
        count_defects=0
        handBool=False

        hand,handBool,maxContourOfHand= humanSkin.getHand(frame)

        drawFrameRate()

        appendUI()
        

        handWithHull = np.zeros((480, 640 , 3), np.uint8)
        if handBool==True:


            handOnly = np.zeros((480, 640 ), np.uint8)
            hullOnly = np.zeros((480, 640 ), np.uint8)
            #openCv hull indices  
            #hull=cv2.convexHull(maxContourOfHand,returnPoints=False)
            
            #openCv hull points
            #hullPoints=cv2.convexHull(maxContourOfHand)
            
            
            #  hull points
            hullPoints=np.array(convexHull.graham_scan(maxContourOfHand[:,0,:].tolist()))
            hullPoints=np.reshape(hullPoints,(hullPoints.shape[0],1,hullPoints.shape[1]))
            
            #  hull indices 
            hull=[]
            for i in range(hullPoints.shape[0]):
                index=np.where(np.logical_and.reduce((maxContourOfHand[:,0]==hullPoints[i,0]) , axis = -1))[0]
                hull.append(index)
            hull=np.array(hull)
            hull=np.sort(hull, axis=0)
            
            
            handWithHull= cv2.drawContours(handWithHull, [maxContourOfHand,hullPoints],-1, color)
            handOnly= cv2.drawContours(handOnly, [maxContourOfHand],-1, 255)
            hullOnly= cv2.drawContours(hullOnly, [hullPoints],-1, 255)
            
            #openCv defects
            '''defects = cv2.convexityDefects(maxContourOfHand, hull)
            if type(defects) != type(None) :
                for i in range(defects.shape[0]):
                    s,e,f,d = defects[i,0]
                    start = tuple(maxContourOfHand[s][0])
                    end = tuple(maxContourOfHand[e][0])
                    far = tuple(maxContourOfHand[f][0])

                    # find length of all sides of triangle
                    a = math.sqrt((end[0] - start[0])**2 + (end[1] - start[1])**2)
                    b = math.sqrt((far[0] - start[0])**2 + (far[1] - start[1])**2)
                    c = math.sqrt((end[0] - far[0])**2 + (end[1] - far[1])**2)
                    s = (a+b+c)/2
                    ar = math.sqrt(s*(s-a)*(s-b)*(s-c))
                    #distance between point and convex hull
                    d=(2*ar)/a
                    # apply cosine rule here
                    angle = math.acos((b**2 + c**2 - a**2)/(2*b*c)) * 57
                   
                    if angle <= 90 and d>50 :
                        count_defects += 1
                        cv2.circle(handWithHull, far, 3, [255,0,0], -1)
                    cv2.line(handWithHull,start, end, [0,255,0], 2)
                     '''
            defects =convexHull.convexityDefects(maxContourOfHand, hull)
            if type(defects) != type(None) :
                for i in range(len(defects)):

                    start=tuple(np.array(defects[i][0][0]))
                    end=tuple(np.array(defects[i][1][0]))
                    far=tuple(np.array(defects[i][2][0]))
    
                    count_defects += 1
                    cv2.circle(handWithHull, far, 3, [255,0,0], -1)
                    cv2.line(handWithHull,start, end, [0,255,0], 2)
            
            # decision making based on count_defects , pixelsRatio , areaRatio

            
            
            #CommonPixels= cv2.bitwise_and(handOnly,hullOnly)
            NumberOfWhitePixels=cv2.countNonZero(hullOnly[2:hullOnly.shape[0]-2,2:hullOnly.shape[1]-2])
            NumberOfWhitePixelsInHand=cv2.countNonZero(handOnly[2:handOnly.shape[0]-2,2:handOnly.shape[1]-2])
            if (NumberOfWhitePixelsInHand==0): 
                NumberOfWhitePixelsInHand=-1
                
            #open cv area
            #areaRatio =cv2.contourArea(maxContourOfHand)/cv2.contourArea(hullPoints)
            areaRatio =humanSkin.getContourArea(maxContourOfHand)/humanSkin.getContourArea(hullPoints)
            #print("area "+str(areaRatio))
            pixelsRatio =NumberOfWhitePixels/NumberOfWhitePixelsInHand
            #print("NumberOfPixels "+str(pixelsRatio))
            if count_defects==4 and pixelsRatio >0.02 and pixelsRatio < 0.1 and areaRatio >0.6 and areaRatio <0.75:        
                count_paper+=1        
            if count_defects==3 and pixelsRatio >0.03 and pixelsRatio < 0.11 and areaRatio >0.6 and areaRatio <0.85:
                count_end+=1
            elif count_defects==2 and pixelsRatio >0.02 and pixelsRatio < 0.13 and areaRatio >0.6 and areaRatio <0.78:
                count_spock+=1
            elif count_defects==1 and  pixelsRatio > 0.07 and pixelsRatio <0.2 and areaRatio >0.6 and areaRatio <0.82 :
                count_scissors+=1
            elif count_defects==1 and pixelsRatio >0.1 and pixelsRatio < 0.25 and areaRatio >0.82 and areaRatio <0.87  :
                count_resume+=1
            elif count_defects==1 and  pixelsRatio > 0.1 and pixelsRatio < 0.2 and areaRatio >0.87 : 
                count_lizard+=1
            elif count_defects==0  and pixelsRatio < 0.1 and areaRatio > 0.6 and areaRatio < 0.75:
                count_pause+=1
            elif count_defects==0 and pixelsRatio >=0.13 and pixelsRatio < 0.7 and areaRatio <0.9 : 
                count_start+=1
            elif count_defects==0 and  pixelsRatio > 0.1 and pixelsRatio < 0.25 and areaRatio >0.9 : 
                count_rock+=1            
            if count_paper>=max_count :
                print("paper")
                key="paper"
                count_paper = 0
                count_rock = 0
                count_scissors = 0
                count_lizard = 0
                count_spock = 0
                count_pause =0
                count_resume =0
                count_start =0
                count_end =0
                
            elif count_end>=max_count :
                print("End")
                key="End"
                count_paper = 0
                count_rock = 0
                count_scissors = 0
                count_lizard = 0
                count_spock = 0
                count_pause =0
                count_resume =0
                count_start =0
                count_end =0
                
            elif count_spock>=max_count :
                print("spock")
                key="spock"                
                count_paper = 0
                count_rock = 0
                count_scissors = 0
                count_lizard = 0
                count_spock = 0
                count_pause =0
                count_resume =0
                count_start =0
                count_end =0
                
            elif count_scissors>=max_count :
                print("scissors")
                key="scissors"                
                count_paper = 0
                count_rock = 0
                count_scissors = 0
                count_lizard = 0
                count_spock = 0
                count_pause =0
                count_resume =0
                count_start =0
                count_end =0
                
            elif  count_lizard>=max_count :
                print("lizard")
                key="lizard"  
                count_paper = 0
                count_rock = 0
                count_scissors = 0
                count_lizard = 0
                count_spock = 0
                count_pause =0
                count_resume =0
                count_start =0
                count_end =0
                
            elif count_resume>=max_count :
                print("resume")
                key="resume"              
                count_paper = 0
                count_rock = 0
                count_scissors = 0
                count_lizard = 0
                count_spock = 0
                count_pause =0
                count_resume =0
                count_start =0
                count_end =0
                
            elif count_pause>=max_count :
                print("pause")
                key="pause"            
                count_paper = 0
                count_rock = 0
                count_scissors = 0
                count_lizard = 0
                count_spock = 0
                count_pause =0
                count_resume =0
                count_start =0
                count_end =0
                
            elif count_start>=max_count :
                print("start")
                key="start"      
                count_paper = 0
                count_rock = 0
                count_scissors = 0
                count_lizard = 0
                count_spock = 0
                count_pause =0
                count_resume =0
                count_start =0
                count_end =0
                
            elif count_rock>=max_count:
                print("rock")
                key="rock"  
                count_paper = 0
                count_rock = 0
                count_scissors = 0
                count_lizard = 0
                count_spock = 0
                count_pause =0
                count_resume =0
                count_start =0
                count_end =0
                
                

            drawFrameRate()
            cv2.imshow('hand', handWithHull)
        else:
            cv2.imshow('hand', handWithHull)
            count_paper = 0
            count_rock = 0
            count_scissors = 0
            count_lizard = 0
            count_spock = 0
            count_pause =0
            count_resume =0
            count_start =0
            count_end =0

        cv2.imshow('frame', newFrame)
        exitKey=cv2.waitKey(1) & 0xFF
        if key == "End" or exitKey==ord('q'):
            stopThread=True
            t.join()
            break
            
        if state=="Wait Start Or End":

            if key == "start":
                state="Playing"
                
        if state=="Playing":

            if key == "pause":
                state="Pause"
                
            if key == "rock":
                gesturesUser="rock"  
                gesturesCpu=gestures[random.randrange(5)]
                whoWin()
                state="Pause"
            
            if key == "paper":
                gesturesUser="paper"   
                gesturesCpu=gestures[random.randrange(5)]
                whoWin()
                state="Pause"

            
            if key == "scissors":
                gesturesUser="scissors"  
                gesturesCpu=gestures[random.randrange(5)]
                whoWin()
                state="Pause"

            
            if key == "lizard":
                gesturesUser="lizard"     
                gesturesCpu=gestures[random.randrange(5)]
                whoWin()
                state="Pause"
            
            if key == "spock":
                gesturesUser="spock"     
                gesturesCpu=gestures[random.randrange(5)]
                whoWin()
                state="Pause"
            
                
                
        if state=="Pause":

            if key == "resume":
                state="Playing"
                whoWinString=""
                gesturesUser=""    
                gesturesCpu=""
                
            if key == "start":
                state="Playing"
                whoWinString=""
                gesturesUser=""    
                gesturesCpu=""
                userScore=0
                cpuScore=0


    cap.release()
    cv2.destroyAllWindows()
    

main()
```
